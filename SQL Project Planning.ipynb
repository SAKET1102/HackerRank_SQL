**SELECT Start_Date, MIN(End_Date) <br>
FROM  <br>
    (SELECT Start_Date FROM Projects WHERE Start_Date NOT IN (SELECT End_Date FROM Projects)) a, <br>
    (SELECT end_date FROM PROJECTS WHERE end_date NOT IN (SELECT start_date FROM PROJECTS)) b <br>
where start_date < end_date <br>
GROUP BY start_date  <br>

**select A.hacker_id, A.name,sum(B.score) as total_score from Hackers A,
(select hacker_id,max(score) as score from submissions group by challenge_id, hacker_id) B  <br>
where B.hacker_id = A.hacker_id <br>
group by A.name,A.hacker_id <br>
having total_score <> 0 <br>
order by total_score desc, A.hacker_id;**
